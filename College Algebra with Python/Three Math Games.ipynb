# Three Math Games

### Tasks / Features:
1. **Scatter plot game**
  - Randomly generate points on a graph and the player has to input the (x,y) coordinates
  - For added difficulty, make the graph larger
2. **Algebra practice game**
  - Generate one-step and two-step problems with random integer values and the player has to input the answer
  - Use positive and negative values. For added difficulty, make the numbers larger
3. **Projectile game**
  - Display a "wall" with random height and location. Player has to move sliders to adjust a parabolic path to clear the wall
  - For added difficulty, make a second level where players enter a, b, and c without sliders

### Scatter Plot Game *(Level: Easy)*

```python
import random
import matplotlib.pyplot as plt
import google.colab.output as goutput
import ipywidgets as widgets
from IPython.display import display

MAX_X = 100
MAX_Y = 100
random_x = random.randint(0,MAX_X)
random_y = random.randint(0,MAX_Y)
guesses = 0

fig, ax = plt.subplots()
ax.set_xlim(0,MAX_X)
ax.set_ylim(0,MAX_Y)
ax.set_xticks(range(0,MAX_X,10))
ax.set_yticks(range(0,MAX_Y,10))
ax.set_title('Scatter Plot Game')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True)
ax.scatter(random_x, random_y)
display(fig.show())

label_x = widgets.Label('Input the x coordinate: ')
input_x = widgets.widget_string.Text()
label_y = widgets.Label('Input the y coordinate:')
input_y = widgets.widget_string.Text()

def check_status(_):
  global guesses

  label_output.value = "Wait...."
  if input_x.value == 'q' or input_y.value == 'q':
    label_output.value = f"The Correct Value: [ x: {random_x} ] & [ y: {random_y} ]"
    return

  guesses += 1
  try:
    if random_x == int(input_x.value) and random_y == int(input_y.value):
      label_output.value = f"Correct! You have guessed {guesses} times (â‰§âˆ‡â‰¦)ï¾‰"
      return
    else:
      label_output.value = "Incorrect! Try Again ãƒ½ï¼ˆâ‰§â–¡â‰¦ï¼‰ãƒŽ"
  except:
      label_output.value = "Error! Please try again with valid inputs (Â´ã€‚ï¼¿ã€‚ï½€)"

label_info = widgets.Label("ðŸ’¡ Enter 'q' in one of the input to give up and showing the actual coordinates")
label_output = widgets.Label('Output: ')
btn_submit = widgets.Button(description='Submit')
btn_submit.on_click(check_status)

display(widgets.VBox([label_info, label_x, input_x, label_y, input_y, btn_submit, label_output]))
```

### Algebra Practice Game *(Level: Easy)*

```python
import ipywidgets as widgets
from IPython.display import display
import sympy as sym
import random

def check_answer(user_answer, act_answer, status):
  try:
    if user_answer == str(act_answer):
      status.value = 'Status: Correct (â‰§âˆ‡â‰¦)ï¾‰'
    else:
      status.value = 'Status: Incorrect ï¿£ã¸ï¿£'
  except:
    status.value = 'Status: Input is invalid or error ã„Ÿ( â–”, â–” )ã„'

def surrend(status, act_answer):
  status.value = f'Status: Surrended (â•¥ï¹â•¥) | Correct Answer: {act_answer}'

available_ops = ['+','-','*','/']
get_op = lambda: random.choice(available_ops)
get_int = lambda: random.randint(-100,100)
x = sym.symbols('x')

problem_1 = f"{x} {get_op()} {get_int()}"
problem_2 = f"{x} {get_op()} {get_int()} {get_op()} {get_int()}"
answer_1 = str(sym.solve(sym.Eq(eval(problem_1), 1, evaluate=False, simplify=False), x, simplify=False)[0])
answer_2 = str(sym.solve(sym.Eq(eval(problem_2), 1, evaluate=False, simplify=False), x, simplify=False)[0])
label_prob_1 = widgets.widget_string.Label(f"{problem_1} = 1")
label_prob_2 = widgets.widget_string.Label(f"{problem_2} = 1")
label = widgets.widget_string.Label('Input the answer: ')
space = widgets.widget_string.Label('')
label_status_1 = widgets.widget_string.Label('Status: ')
label_status_2 = widgets.widget_string.Label('Status: ')
input_1 = widgets.widget_string.Text()
input_2 = widgets.widget_string.Text()
submit_1 = widgets.widget_button.Button(description='Submit')
surrend_1 = widgets.widget_button.Button(description='Surrend')
submit_2 = widgets.widget_button.Button(description='Submit')
surrend_2 = widgets.widget_button.Button(description='Surrend')

submit_1.on_click(lambda _: check_answer(input_1.value, answer_1, label_status_1))
surrend_1.on_click(lambda _: surrend(label_status_1, answer_1))
submit_2.on_click(lambda _: check_answer(input_2.value, answer_2, label_status_2))
surrend_2.on_click(lambda _: surrend(label_status_2, answer_2))

display(
  widgets.VBox([label_prob_1, label, input_1, label_status_1, widgets.HBox([submit_1, surrend_1]), space]),
)
display(
  widgets.VBox([label_prob_2, label, input_2, label_status_2, widgets.HBox([submit_2, surrend_2])])
)
```

### Projectile Game *(Level: Super Very Easy)*



```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
import random
import numpy as np

MAX_X = 100
MAX_Y = 100
MAX_INPUT = 20

vx = 0
vy = 0
wx = random.randint(MAX_X//4,MAX_X//1.5)
wy = random.randint(0,MAX_Y//2)
x = np.linspace(0, MAX_X, 1000)

def f(a: float|int, b: int, c: int):
  y = a*x**2 + b*x + c
  collided = (a*wx**2 + b*wx + c) <= wy

  fig, ax = plt.subplots()
  ax.set_xlim(0,MAX_X)
  ax.set_ylim(0,MAX_Y)
  ax.set_xticks(range(0,MAX_X,10))
  ax.set_yticks(range(0,MAX_Y,10))
  ax.set_title('Projectile Game')
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.grid(True)
  ax.plot([wx,wx],[0,wy],'b')
  ax.plot(x,y,'r' if collided else 'g')
  fig.show()

interactive_plot = widgets.interactive(f, a=(-0.5, 0, 0.01), b=(0,MAX_INPUT,1), c=(0,MAX_INPUT,1))
for child in interactive_plot.children:
  child.value = 0
  child.layout.width = '750px'
display(interactive_plot)
```

### Projectile Game *(Level: A Little More Difficult)*



```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
import google.colab.output as goutput
import random
import numpy as np

MAX_X = 100
MAX_Y = 100
MAX_INPUT = 10

vx = 0
vy = 0
wx = random.randint(MAX_X//4,MAX_X//1.5)
wy = random.randint(0,MAX_Y//2)
x = np.linspace(0, MAX_X, 1000)

label_a = widgets.Label('Coefficient a (-infinity - 0): ')
input_a = widgets.widget_string.Text()
label_b = widgets.Label('Coefficient b (0 - infinity):')
input_b = widgets.widget_string.Text()
label_c = widgets.Label('Coefficient c (0 - infinity):')
input_c = widgets.widget_string.Text()
label_status = widgets.Label('Status: ')

def f(a: float|int, b: int, c: int):
  y = a*x**2 + b*x + c
  collided = (a*wx**2 + b*wx + c) <= wy
  fig, ax = plt.subplots()

  ax.set_xlim(0,MAX_X)
  ax.set_ylim(0,MAX_Y)
  ax.set_xticks(range(0,MAX_X,10))
  ax.set_yticks(range(0,MAX_Y,10))
  ax.set_title('Projectile Game')
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.grid(True)
  ax.plot([wx,wx],[0,wy],'b')
  ax.plot(x,y,'r' if collided else 'g')
  display(fig.show())

  return not collided


def plot_figure(widget):
  try:
    a = float(input_a.value) if input_a.value else 0
    b = float(input_b.value) if input_b.value else 0
    c = float(input_c.value) if input_c.value else 0
    is_valid = a <= 0 and b >= 0 and c >= 0
  except:
    a, b, c = 0, 0, 0
    is_valid = False

  goutput.clear()
  display(widgets.VBox([label_a, input_a, label_b, input_b, label_c, input_c, btn_submit, label_status]))
  is_correct = f(a, b, c)

  if not is_valid:
    label_status.value = 'Status: Invalid Input'
    return

  label_status.value = 'Status: Correct' if is_correct else 'Status: Inorrect'

btn_submit = widgets.Button(description='Submit')
btn_submit.on_click(plot_figure)

is_correct = f(0, 0, 0)
display(widgets.VBox([label_a, input_a, label_b, input_b, label_c, input_c, btn_submit, label_status]))
```
